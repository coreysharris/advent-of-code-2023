```python
import re

def ints_in_line(s):
    return list(map(int, re.findall(r"\d+", s)))
```

```python
filename = "input.txt"
# filename = "example.txt"

with open(filename, "r") as f:
    input = f.readlines()

instructions = input[0].strip()
print(instructions)
places = list(map(lambda s: re.findall(r"[0-9A-Z]{3}", s), input[2:]))
# print(places[:4])
places
```

# Part 1

```python
map_ = {}
for place in places:
    map_[place[0]] = {
        'L': place[1],
        'R': place[2],
    }
```

```python
from itertools import cycle

current = 'AAA'
steps = 0
for instr in cycle(instructions):
    current = map_[current][instr]
    steps += 1
    if current == 'ZZZ':
        break
print(steps)
```

# Part 2

```python
from itertools import cycle

print(len(instructions))
starts = [a for a in map_.keys() if a.endswith('A')]
print(starts)
# current = 'AAA' 
# current = 'GPA'
current = 'GTA'
step_tracker = []
steps = 0
for start in starts:
    current = start
    steps = 0
    for instr in cycle(instructions):
        current = map_[current][instr]
        steps += 1
        if current.endswith('Z'):
            print(steps, current, steps % len(instructions))
            step_tracker.append(steps)
            break
```

```python
from math import lcm

lcm(*step_tracker)
```
