```python
import re

def ints_in_line(s):
    return list(map(int, re.findall(r"\d+", s)))
```

```python
filename = "input.txt"
# filename = "example1.txt"

with open(filename, "r") as f:
    input = f.readlines()

lines = [l.split() for l in input]
for line in lines:
    line[1] = tuple(ints_in_line(line[1]))
print(lines)
# maze = [[c for c in s.strip()] for s in input]
# print(maze)
```

```python
from functools import cache

@cache
def num_arrangements(chart, groups):
    if len(groups) == 0 and '#' not in chart:
        return 1
    if len(chart) == 0:
        return 0
    if chart[0] == '.':
        return num_arrangements(chart[1:], groups)
    if chart[0] == '#':
        g1 = groups[0]
        if len(groups) == 1:
            if chart.replace('?', '#').startswith("#" * g1) and "#" not in chart[g1:]:
                return 1
            else:
                return 0
        elif chart.replace('?', '#').startswith("#" * g1) and (len(chart) == g1 or chart[g1] in (".", "?")): 
            return num_arrangements(chart[g1+1:], groups[1:])
        else:
            return 0
    if chart[0] == '?':
        return num_arrangements("#" + chart[1:], groups) + num_arrangements("." + chart[1:], groups)
```

```python
def solution(lines):
    arrangement_counts = []
    for line in lines:
        arrangement_counts.append(num_arrangements(*line))
        # print(line, arrangement_counts)
    return sum(arrangement_counts)
```

```python
solution(lines)
```
