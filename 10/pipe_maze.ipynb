# Part 1

```python
mazed = dict()
num_rows = len(maze)
num_cols = len(maze[0])
for r, row in enumerate(maze):
    for c, val in enumerate(row):
        mazed[(r,c)] = val
        if val == 'S':
            start = (r,c)

print(start)
```

```python
dirs = [(0,1), (0,-1), (1,0), (-1,0)]
def neighbors(node):
    x,y = node
    for dx, dy in dirs:
        if 0 <= x+dx < num_rows and 0 <= y+dy < num_cols:
            yield ((x+dx, y+dy), (dx, dy))
```

```python
from collections import deque

SOUTH = ['J', 'L', '|']
NORTH = ['F', '7', '|']
EAST = ['J', '7', '-']
WEST = ['F', 'L', '-']

connections = {
    (0,1):  (['S'] + WEST,  EAST),
    (0,-1): (['S'] + EAST,  WEST),
    (1,0):  (['S'] + NORTH, SOUTH),
    (-1,0): (['S'] + SOUTH, NORTH),
}
def valid_connection(curr, next, dir):
    return curr in connections[dir][0] and next in connections[dir][1]

def run_maze(maze):
    visited = []
    Q = deque([start])

    while Q:
        node = Q.pop()
        for nbhr, nbhr_dir in neighbors(node):
            if nbhr in visited:
                continue
            if valid_connection(maze[node], maze[nbhr], nbhr_dir):
                Q.appendleft(nbhr)
        visited.append(node)
    return visited
```

```python
loop = run_maze(mazed)
print((len(loop)-1) // 2)
```

```python
from collections import deque

RIGHT = '@'
LEFT = '%'

def maybe_label(maze, node, label):
    # print(f"maybe label {node} (={maze[node]}) with {label}:", sep=' ')
    if maze.get(node, 'X') == '.':
        # print(f"LABELED {node} (={maze[node]}) with {label}:", sep=' ')
        maze[node] = label

def run_maze(maze):
    maze = {k:v for k,v in maze.items()}
    visited = []
    Q = deque([start])
    while Q:
        node = Q.pop()
        for nbhr, nbhr_dir in neighbors(node):
            if nbhr in visited:
                continue
            nr, nc = nbhr
            r, c = node
            if nbhr_dir == (1,0) and maze[nbhr] in SOUTH and maze[node] in ['S'] + NORTH:
                for N in [(nr, nc-1), (r, c-1)]:
                    maybe_label(maze, N, RIGHT)
                for N in [(nr, nc+1), (r, c+1)]:
                    maybe_label(maze, N, LEFT)
                Q.appendleft(nbhr)
                break
            elif nbhr_dir == (-1,0) and maze[nbhr] in NORTH and maze[node] in ['S'] + SOUTH:
                for N in [(nr, nc-1), (r, c-1)]:
                    maybe_label(maze, N, LEFT)
                for N in [(nr, nc+1), (r, c+1)]:
                    maybe_label(maze, N, RIGHT)
                Q.appendleft(nbhr)
                break
            elif nbhr_dir == (0,1) and maze[nbhr] in EAST and maze[node] in ['S'] + WEST:
                for N in [(nr+1, nc), (r+1, c)]:
                    maybe_label(maze, N, RIGHT)
                for N in [(nr-1, nc), (r-1, c)]:
                    maybe_label(maze, N, LEFT)
                Q.appendleft(nbhr)
                break
            elif nbhr_dir == (0,-1) and maze[nbhr] in WEST and maze[node] in ['S'] + EAST:
                for N in [(nr+1, nc), (r+1, c)]:
                    maybe_label(maze, N, LEFT)
                for N in [(nr-1, nc), (r-1, c)]:
                    maybe_label(maze, N, RIGHT)
                Q.appendleft(nbhr)
                break
        visited.append(node)
    return maze
```

```python
lr_maze = run_maze(cleared_maze)
```

```python
def label_fill(maze, labels):
    maze = {k:v for k,v in maze.items()}
    change_count = 1
    while change_count > 0:
        change_count = 0
        for k,v in maze.items():
            if v in labels:
                for nbhr, _ in neighbors(k):
                    if maze[nbhr] == '.':
                        maze[nbhr] = v
                        change_count += 1
    return maze

def get_label_counts(maze):
    right_count = 0
    left_count = 0
    for v in maze.values():
        if v == RIGHT:
            right_count += 1
        if v == LEFT:
            left_count += 1
    return right_count, left_count
```

```python
labeled_maze = label_fill(lr_maze, [RIGHT, LEFT])
print(get_label_counts(labeled_maze))
```

```python
def label_inside_space(maze):
    inside_count = 0
    for r in range(num_rows):
        inside = False
        for c in range(num_cols):
            if maze[(r,c)] in 'J|':
                inside = not inside
            if inside and maze[(r,c)] == '.':
                inside_count += 1
                maze[(r,c)] = '#'
    return inside_count
```

```python
def add_delta(node, dir):
    return node[0]+dir[0], node[1]+dir[1]

RIGHT = '@'
LEFT = '%'
periphery = {
    (0,1):  {LEFT: (-1,0), RIGHT: (1, 0)}, # right
    (0,-1): {LEFT: (1,0), RIGHT: (-1, 0)}, # left
    (1,0):  {LEFT: (0,1), RIGHT: (0,-1)},  # down
    (-1,0): {LEFT: (0,-1), RIGHT: (0,1)},  # up
}

def maybe_label(maze, nodes, dir):
    for node in nodes:
        for label, delta in periphery[dir].items():
            candidate = add_delta(node, delta)
            if maze.get(candidate, 'X') == '.':
                maze[candidate] = label

def run_maze2(maze):
    maze = {k:v for k,v in maze.items()}
    visited = []
    Q = deque([start])

    while Q:
        node = Q.pop()
        for nbhr, nbhr_dir in neighbors(node):
            if nbhr in visited:
                continue
            if valid_connection(maze[node], maze[nbhr], nbhr_dir):
                maybe_label(maze, [node, nbhr], nbhr_dir)
                Q.appendleft(nbhr)
                break
        visited.append(node)
    return maze
```
